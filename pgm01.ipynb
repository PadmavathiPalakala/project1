```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#when was the property sold in yr and month
df.groupby(['YrSold', 'MoSold']).Id.count().plot(kind ='bar',figsize=(14,4))
plt.title('which year and month property was sold')
plt.show()
```

```python
#where is the property location
df.groupby(['Neighborhood']).Id.count().plot(kind ='bar',figsize=(10,4))
plt.title('where is the property located')
plt.show()
```

```python
#distribution of numerical features
f = pd.melt(df,value_vars = sorted(numeric_data))
f
```

```python
g = sns.FacetGrid(f, col='variable',col_wrap =4,sharex=False,sharey=False)
g =g.map(sns.distplot, 'value')
```

```python
#conversion of numeric to categorical features
df['MSSubClass'] = df.MSSubClass.apply(lambda x:str(x))
```

```python
df['YrSold'] = df.MSSubClass.apply(lambda x:str(x))
```

```python
df['MoSold'] = df.MSSubClass.apply(lambda x:str(x))
```

```python
numeric_data = df.select_dtypes(include = [np.number])
categorical_data = df.select_dtypes(exclude = [np.number])
print('there are {0} numerical and {1} categorical features in the data '.\
     format(numeric_data.shape[1], categorical_data.shape[1]))
                                                                   # to check 
```

```python
#plot count of categorical features
f = pd.melt(df,value_vars = sorted(categorical_data))
g = sns.FacetGrid(f, col='variable',col_wrap=4,sharex=False,sharey=False)
plt.xticks(rotation='vertical')
g =g.map(sns.countplot, 'value')
[plt.setp(ax.get_xticklabels(), rotation=60) for ax in g.axes.flat]
g.fig.tight_layout()
plt.show()
```
